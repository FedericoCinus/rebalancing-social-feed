```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv("./scripts/prova_heat.csv")
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 2.75))
methods = ["LcGD", "SCS",]
cmap = matplotlib.cm.get_cmap('RdYlGn_r')
for i, ax in enumerate(axs):
    means = df[df.method==methods[i]].pivot_table("OBJ/UB", "", "Number of nodes", aggfunc='mean')
    sb.heatmap(means, ax=ax, cmap=cmap, vmin=.5, vmax=1., annot=True, linewidth=.5, fmt=".0%")
    ax.set_title(f"Method: {methods[i]}")
```
